```python
import numpy as np
string=input("Enter plaintext:")
key=int(input("ENTER KEY:"))
m=np.zeros((key,len(string)),dtype="object")

dir=1
row=0
for j in range(len(string)):
  if dir==1 and row<=key-1:
    m[row][j]=string[j]
    row+=1
    if row==key-1:
      dir=0
  elif dir==0 and row>=0:
    m[row][j]=string[j]
    row-=1
    if row==0:
      dir=1
print(m)
cipher=""
for i in range(key):
  for j in range(len(string)):
    if m[i][j]!=0:
      cipher+=m[i][j]
print(cipher)
```
