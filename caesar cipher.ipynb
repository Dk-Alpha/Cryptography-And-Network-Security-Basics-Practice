```python
def caesar_cipher(text, shift):
    result = ""

    # Iterate through each character in the text
    for char in text:
        # Check if the character is an uppercase letter
        if char.isupper():
            # Apply the shift and wrap around the alphabet
            new_char = chr((ord(char) + shift - 65) % 26 + 65)
            result += new_char
        # Check if the character is a lowercase letter
        elif char.islower():
            # Apply the shift and wrap around the alphabet
            new_char = chr((ord(char) + shift - 97) % 26 + 97)
            result += new_char
        # Otherwise, leave the character as it is
        else:
            result += char

    return result

#made by Dhruval Patel

text = "Dhruval is Jod"
shift = 3
encrypted_text = caesar_cipher(text, shift)
print("Original Text is:",text)
print("Encrypted Text is: ",encrypted_text)
```
