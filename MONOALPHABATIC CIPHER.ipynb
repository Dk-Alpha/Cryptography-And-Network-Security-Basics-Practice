```python
pt1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
ci=['z','y','x','w','v','u','t','s','r','q','p','o','n','m','l','k','j','i','h','g','f','e','d','c','b','a']  #any random letter can be given
#made by Dhruval Patel
plaintext=input("Enter a text:")
ciphertext=""
for i in range(len(plaintext)):
  j=pt1.index(plaintext[i])
  ciphertext+=ci[j]
print(ciphertext)

plaintext=""
for i in range(len(ciphertext)):
  j=ci.index(ciphertext[i])
  plaintext+=pt1[j]
print(plaintext)
```
